```python
df_order_review = pd.read_csv('order_reviews_dataset.csv')
df_order_review.head()
```

```python
df_order = pd.read_csv('orders_dataset.csv')
df_order.head()
```

```python
df_product_name = pd.read_csv('product_category_name_translation.csv')
df_product_name.head()
```

```python
df_product = pd.read_csv('products_dataset.csv')
df_product.head()
```

```python
df_seller = pd.read_csv('sellers_dataset.csv')
df_seller.head()
```

```python
df_geolocation = pd.read_csv('geolocation_dataset.csv')
df_geolocation.head()
```

TRANSFORM

```python
df_full_order = df_order.merge(df_order_item,on='order_id',how='inner').merge(df_order_payment, on='order_id',how='inner').merge(df_order_review, on='order_id',how='inner')
df_full_order.head(5)
```

```python
# show missing value
df_full_order[df_full_order.isna().any(axis=1)]
```

```python
# duplicate data
df_full_order[df_full_order.duplicated()]
```

```python
df_full_order = df_full_order.dropna(axis=1)
```

```python
# show missing value
df_full_order[df_full_order.isna().any(axis=1)]
```

```python
df_full_order.head(10)
```

```python
df_full_order.loc[10]
```

```python
df_full_order=df_full_order.rename(columns={'order_id':'ID Order', 'customer_id':'ID Customer', 'order_status':'Status', 'order_purchase_timestamp':'Timestamp Order', 'order_estimated_delivery_date':'Estimated Delivery', 'order_item_id':'ID Item', 'product_id':'ID Product', 'seller_id':'ID Seller', 'shipping_limit_date':'Shipping Limit', 'price':'Price', 'freight_value':'Freight', 'payment_sequential':'Sequential', 'payment_type':'Type', 'payment_installments':'Installments', 'payment_value':'Payment', 'review_id':'ID Review', 'review_score':'Score', 'review_creation_date':'Creation Date', 'review_answer_timestamp':'Timestamp Answer '})
df_full_order.head()
```

```python
df_full_order.to_csv('full_Order.csv')
```

```python
df_full_product = pd.merge(df_product, df_product_name, on='product_category_name', how='inner')
df_full_product.head()
```

```python
# show missing value
df_full_product[df_full_product.isna().any(axis=1)]
```

```python
df_full_product = df_full_product.dropna(axis=1)
df_full_product[df_full_product.duplicated()]
```

```python
df_full_product.head(10)
```

```python
df_full_product.loc[1]
```

```python
df_full_product=df_full_product.rename(columns={'product_id':'ID Product', 'product_category_name':'Product Category', 'product_name_lenght':'Lenght Product Name', 'product_description_lenght':'Description Lenght', 'product_photos_qty':'Photos Qty', 'product_category_name_english':'Category English Name'})
df_full_product.head()
```

```python
df_full_product.to_csv('full_Product.csv')
```

```python
df_seller_geo_customer = df_customer.merge(df_seller, left_on='customer_zip_code_prefix', right_on='seller_zip_code_prefix', how='inner').merge(df_geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
df_seller_geo_customer.head(5)
```

```python
# show missing value
df_seller_geo_customer[df_seller_geo_customer.isna().any(axis=1)]
```

```python
df_seller_geo_customer[df_seller_geo_customer.duplicated()]
```

```python
# drop duplicate
df_seller_geo_customer = df_seller_geo_customer.drop_duplicates()
df_seller_geo_customer[df_seller_geo_customer.duplicated()]
```

```python
df_seller_geo_customer=df_seller_geo_customer.rename(columns={'customer_id':'ID Customer', 'customer_unique_id':'ID Uniqe Customer', 'customer_zip_code_prefix':'Zip Code Customer', 'customer_city':'Customer City', 'customer_state':'Customer State', 'seller_id':'ID Seller','seller_zip_code_prefix':'Zip Code Seller','seller_city':'Seller City','seller_state':'Seller State','geolocation_zip_code_prefix':'Zip Code Geolocation','geolocation_lat':'Geolocation Lat','geolocation_lng':'Geolocation Ing','geolocation_city':'Geolocation City','geolocation_state':'Geolocation State'})
df_seller_geo_customer.head()
```

```python
df_seller_geo_customer.to_csv('full_SelgeTomer.csv')
```
